## Abstract

A major problem for large airlines are flight delays, which can lead to costs to account for rescheduling and customer churn. Being able to reliably identify scenarios that can lead to flight delays and accurately predicting delay time can aid these major airlines in being ahead of the curve. In order to achieve this, our project utilizes a regression modeling approach to predict flight delay times based on a set of airline and weather related features. 

For our analysis, we utilize flight delay data from the U.S. Department of Transportation TransStats dataset and weather data from the National Oceanic and Atmospheric Administration repository. We limit our initial EDA to 1 quarter (3 months) of data from 2022. From our EDA, we found a heavy class imbalance between the potential classification outcomes, with about 83% having departure delays less than 15 mins (excluding NA values). This informed our decision to choose a regression approach as opposed to a classification approach.  Our end to end data pipeline will begin with a data ingestion and joining step, where we consider relevant weather and flight related features for departure delay prediction. We choose to do a custom join to get 12 months of data on both flights and weather.  The data is then validated and cleaned, where all cancelled flight data is removed, date-related features are encoded into strings, and features with large amounts of NA values are cleaned using medians. We then run a training step with multiple model approaches, including linear regression (both regularized and non-regularized), gradient boosted tree regression, and multilayer perceptron. We then evaluate the model approaches on 6 specific metrics: root mean squared error (RMSE), mean absolute error (MAE), R2, on-time performance accuracy (OTPA), same-day delay rate (SDDR), and F1. 

We will use time-series based cross validation for our training/test/validation splits to avoid data leakage. We will continuously fine-tune the model pipeline based on evaluation performance.

##Train/Test/Validation Split


When designing our train/val/test split, 3 key rationales guided our design:

**Seasonality and Drift**


Flight behavior in October is much different than December, which can be explained by extreme holiday schedules and higher frequencies of winter storms. We want to use December as our held out, test set. If the model performs well on the high stress conditions, then we can confidently say it has generalized well from learning on the October and November training sets


**Data Leakage**


Due to the temporal nature of the dataset, it is critical that the model is forbidden from "looking ahead", i.e. using data from the future to make a prediction on a past event during training. Splitting the data into time series will prevent the model from using this source of data leakage, as it can only use past information to learn.


**Feature Normalization**


All scaling parameters will be calculated only on the training set and then applied to the validation and test splits. This will prevent data leakage during preprocessing and ensure that relative "future" data will not influence "past" features.


**Time-Series Cross Validation**

The train/val/test will be maintained chronologically using a time-series cross validation strategy, ensuring that the model is always tested on data that occurs after the training set. In addition, instead of a single split, training data will grow over time using multiple folds.

Fold 1: Train on Oct; Validate on Nov (Week 1).

Fold 2: Train on Oct–Nov (Week 1); Validate on Nov (Week 2).

Final Evaluation: Train on Oct–Nov; Test on Dec (Holiday Peak).

| Split | Data Period | Purpose |
| :--- | :--- | :--- |
| Training Set | October 1 – November 15 | Used to fit model parameters and learn baseline weather effects. |
| Validation Set | November 16 – November 30 | Used for hyperparameter tuning (e.g., alpha in Lasso, depth of GBT) and stacking weight optimization. |
| Hold-out Test Set | December 1 – December 31 | A held out dataset used only for final evaluation to measure performance during high-variance holiday peaks. |

## Machine Algorithms and Metrics



Based on our EDA, we decided to frame the flight delay prediction problem as a regression task, with the target variable the continuous departure delay in minutes (DEP_DELAY). Two key factors derived from our EDA philosophically drove this decision:

**Granularity and Operational Utility:** The operational cost of a 20 minute delay is wildly different from a 3 hour delay; approaching this problem with classification would require binning these two delays as the same, despite the operational impacts the 3 hour delay would have. Regression allows for more precision in logistical planning and passenger notification since its output can be binned into separate buckets, giving flexibility for a secondary analysis

**EDA:** The distribution of delays is highly skewed (Median: -1 min, Mean: 10.4 min), with extreme outliers reaching 2,710 minutes. Regression is better suited to modelling this heavy right skewed distribution compared to classification.



## Proposed Algorithms

**Regularized Linear Regression (Lasso, Ridge, and Elastic Net)**

Implementation: pyspark.ml.regression.LinearRegression

Loss Function: Mean Squared Error (MSE) with L1/L2 penalties.

$$\text{Loss} = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2 + \lambda_1 \sum |w_j| + \lambda_2 \sum w_j^2$$

Rationale: Our EDA revealed high dimensionality due to categorical features like ORIGIN_AIRPORT_ID and OP_CARRIER, as well as multicollinearity among weather features (e.g., HourlyVisibility and HourlyPrecipitation). Lasso (L1) will serve as an automated feature selection tool to "zero out" irrelevant predictors, while Ridge (L2) will prevent overfitting to noisy weather data.

**Gradient Boosted Trees**

Implementation: pyspark.ml.regression.GBTRegressor

Loss Function: MSE + Tree Reg.

Rationale: Weather impacts on aviation are often non-linear, such as wind speeds that have a negligible effect until it crosses a threshold at which point delays increase exponentially. GBTs are highly effective at capturing these complex, threshold-gated interactions and handling the skewed feature distributions (DailySnowfall) identified in our EDA.

**Multilayer Perceptron for Regression**

Implementation: pyspark.ml.regression.DecisionTreeRegressor

Loss Function: MSE (Backpropagation)

Rationale: Given the temporal and network-based nature of flight data, adding deep learning may help capture hidden patterns present between airport congestion factors and cascading delays that a linear regression model might otherwise miss.


**Stacking Ensemble**

Implementation: Custom PySpark Pipeline

Loss Function: Ridge Regression

Rationale: Implementing a layer to intelligently weigh model outputs will help minimize global errors as the layer learns when to trust each model, reducing bias and variance and stronger results than any individual model.

### Evaluation Metrics

**Root Mean Squared Error (RMSE)** 

RMSE is our primary metric since it will penalize larger error more harshly, critical for the problem context, where airlines will face massive backlash the larger the miscalculation in delay times.

$$\text{RMSE} = \sqrt{\frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2}$$

**Mean Absolute Error (MAE)**

After identifying extreme outliers in our target variable, DEP_DELAY, MAE will provide a measure of the typical error in minutes without being overly influenced by weather anomalies.

$$\text{MAE} = \frac{1}{n} \sum_{i=1}^{n} |y_i - \hat{y}_i|$$

**Coefficient of Determination (R^2)**

R^2 will indicate the proportion of variance in DEP_DELAY that our combined flight and weather features can explain.

$$R^2 = 1 - \frac{\sum (y_i - \hat{y}_i)^2}{\sum (y_i - \bar{y})^2}$$

**On-Time Performance Prediction Accuracy (OTPA)**

 Accuracy in correctly identifying if a flight meets the DOT 15-minute threshold, treating prediction as a binary outcome, either on time or delayed. 

 TP = Correct predictions for On Time
 TN = Correct predictions for Delayed

 $$OTPA = \frac{TP_{on\_time} + TN_{on\_time}}{TP_{on\_time} + TN_{on\_time} + FP_{on\_time} + FN_{on\_time}}$$

**Severe Delay Detection Rate (SDDR)**
The recall measure for the percentage of major delays (60min>) that were correctly identified, this measure serves as secondary evaluation metric to gauge how many critical errors the model(s) made.

$$SDDR = \frac{TP_{severe}}{TP_{severe} + FN_{severe}}$$

**F1 Score**
F1 will serve as the weighted metric of precision and recall, gauging the models balance of being correct in predicting delays, and predicting all possible delays.

$$F1_{delayed} = 2 \times \frac{\text{Precision}_{delayed} \times \text{Recall}_{delayed}}{\text{Precision}_{delayed} + \text{Recall}_{delayed}}$$

$$\text{Precision}_{delayed} = \frac{TP_{delayed}}{TP_{delayed} + FP_{delayed}}$$

$$\text{Recall}_{delayed} = \frac{TP_{delayed}}{TP_{delayed} + FN_{delayed}}$$

### Machine Learning Pipeline

### Data Ingestion

The pipeline draws from three sources: flight records containing operational data (departure times, origin/destination, delay values), hourly and daily weather observations keyed to a STATION identifier, and a station lookup table that serves as the critical bridge between the two - mapping each station_id to a neighbor_id corresponding to a nearby airport.

### Data Join
The station lookup table bridges the flight and weather datasets by resolving each flight's ORIGIN_AIRPORT_ID to a neighbor_id, which corresponds to a STATION identifier in the weather observations. Once that link is established, each flight record is matched to the weather reading closest to its scheduled departure by rounding FL_DATE + CRS_DEP_TIME to the nearest hour and aligning it with the DATE field in the weather table. The result is a single unified record per flight containing both operational and atmospheric data.

### Data Validation

Before any transformations are applied, the merged dataset is checked for quality issues. Key checks include verifying that the target variable DEP_DELAY is non-null across all training records, confirming that fields like HourlyWindSpeed and HourlyVisibility fall within physically plausible ranges, scanning for duplicate records on the composite key of FL_DATE + TAIL_NUM + CRS_DEP_TIME, and flagging any stations with disproportionately high missing rates across weather columns.

### Data Cleaning

Issues surfaced during validation are resolved here. Flights where CANCELLED = 1 are dropped entirely, as they carry no departure delay value and represent a fundamentally different prediction problem. Sparse weather fields such as HourlyPrecipitation and DailySnowfall are imputed using median values grouped by station and month. The DATE and FL_DATE timestamp fields are cast to a consistent type, and the raw encoded HourlySkyConditions strings (e.g. OVC:08 89) are parsed into numeric cloud cover values.

### Feature Engineering

The feature vector is standardized for model ingestion. Continuous features such as HourlyWindSpeed and DISTANCE are scaled using StandardScaler. Nominal categoricals - OP_CARRIER, ORIGIN_STATE_ABR, and DEP_TIME_BLK - are one-hot encoded. All resulting columns are assembled into a single dense vector using PySpark's VectorAssembler, which serves as the uniform input format across all three base models.

### Feature Preprocessing

The feature vector is standardized for model ingestion. Continuous features such as HourlyWindSpeed and DISTANCE are scaled using StandardScaler. Nominal categoricals - OP_CARRIER, ORIGIN_STATE_ABR, and DEP_TIME_BLK - are one-hot encoded. All resulting columns are assembled into a single dense vector using PySpark's VectorAssembler, which serves as the uniform input format across all three base models.

### Model Training

Three base learners are trained in parallel on the preprocessed feature vector: regularized linear regression (Lasso, Ridge, and Elastic Net), a Gradient Boosted Tree Regressor, and a Multilayer Perceptron. Their out-of-fold predictions are then passed to a Ridge regression meta-learner, which learns to optimally weight each model's output into a final prediction. Details on algorithm rationale and loss functions are covered in the Algorithms and Metrics section.

Stacking Ensemble collects the out-of-fold predictions from all three base models and feeds them as input to a Ridge regression meta-learner, which learns optimal weights for combining the three outputs into a final prediction.

### Evaluation
The final ensemble is scored against a held-out test set using six metrics: RMSE, MAE, R², OTPA, SDDR, and F1. Full definitions and equations for each metric are provided in the Algorithms and Metrics section.
